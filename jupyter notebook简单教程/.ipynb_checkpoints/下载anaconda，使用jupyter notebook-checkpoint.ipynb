# 下载anaconda
1. anaconda是python的包管理器、环境管理器
2. 下载链接 https://www.anaconda.com/download/
3. 用清华镜像下载速度较快 https://mirror.tuna.tsinghua.edu.cn/anaconda/archive/

# 使用jupyter noteboot
## 打开jupyter noteboot两种方式
1. 在开始菜单anaconda目录下打开jupyter notebook，文件默认路径：C:\Users\Administrator
2. cmd中打开，cd到指定文件目录，输入命令jupyter notebook，前提是安装是勾选了加入环境变量
