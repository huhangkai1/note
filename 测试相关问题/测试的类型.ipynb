# 测试的类型
## 1. 功能测试
    1. 链接测试。链接是否正确跳转，是否存在空页面和无效页面，是否有不正确的出错信息返回。
    2. 提交功能的测试。
    3. 多媒体元素是否可以正确加载和显示。
    4. 多语言支持是否能够正确显示选择的语言等。
## 2. 易用性测试
    1. 界面测试
    2. 页面是否风格统一，美观
    3. 页面布局是否合理，重点内容和热点内容是否突出
    4. 控件是否正常使用
    5. 对于必须但未安装的控件，是否提供自动下载并安装的功能
    6. 完全是站在用户的角度上进行的测试。
## 3. 性能测试
    1. 负载测试的目的是：确认系统在正常指标下的最大负载。
    2. 压力测试，测试出系统所能承受的最大极限。
    3. 稳定性测试一般是稍大于业务量的一个负载，对系统进行的一个持续的，长时间的测试，确定系统在较长运行时间的情况下，系统的稳定性情况
    4. 性能测试很重要的一个工作就是性能分析、性能评价，也是用户验收的一个重要指标。
## 4. 数据库测试
    1. 要具体决定是否需要开展。数据库一般需要考虑连结性，对数据的存取操作，数据内容的验证等方面。
## 5. 安全测试
    1. 对产品进行检验以验证产品符合安全需求定义和产品质量标准的过程 。特别是一些安全要求较高的产品，比如登录、注册功能等。
    2. 是否存在溢出错误，导致系统崩溃或者权限泄露
    3. 相关开发语言的常见安全性问题检查，例如SQL注入等
## 6. 兼容性测试
    1. 软件本身的兼容性，是指对历史版本的数据、功能等进行兼容
    2. 不同平台下的兼容，软件可能运行在多个平台上，比如LINUX等有很多版本，在这些平台上都需要验证
    3. 操作系统的兼容性，比如android，有多种手机。
    4. 浏览器的兼容性
    5. 数据库的兼容性
## 7. 安装测试
    1. 确保该软件在正常情况和异常情况的不同条件下，例如，进行首次安装、升级、完整的或自定义的安装都能进行安装；异常情况包括磁盘空间不足、缺少目录创建权限、安装过程中关机重启等。
## 8. 文档测试
    1. 档测试是检验样品用户文档的完整性、正确性、一致性、易理解性、易浏览性。包括用户手册、使用说明、用户帮助文档等
## 9. 本地化测试
    1. 对不同地区的用户推出不同的版本、比如中文版、英文版等。
## 10. 可靠性测试
    1. 这个主要是硬件方面的，比如高低温测试、防水防尘等测试
## 11. 竞品测试
    1. 测试竞争产品的功能、性能等，和自己的产品的一个对比
## 12. 体验测试
    1. 体验一下产品开发的如何，就会去体验一下，一般是在开发环境下体验，如果产品规模较小，也可能直接拿回家体验
## 13. 无障碍测试
    1. 主要是针对特殊人群，比如老年人、身体有残疾的用户
