# 如何指定测试计划
https://jingyan.baidu.com/article/cb5d6105243ade005c2fe0fa.html
## 1. 确定测试策略
    1. 确定测试阶段（单元测试、集成测试和系统测试）
    2. 要执行的测试类型（功能测试、性能测试、负载测试、强度测试等）
## 2. 确定测试系统
    1. 确定测试环境
    2. 确定测试工具
    3. 确定配置情况
    4. 确定测试资源
    5. 测试人力资源
    6. 测试非人力资源（计算机、工具等）
## 3. 确定测试任务
    1. 根据本阶段测试需求，细化测试任务
    2. 划分任务优先级，和主要任务关联关系
    3. 确定资源情况
## 4. 评估和确定测试工作量
    1. 统计和分析历史数据
        分析以前的同类项目
        同行专家判断
        分解细化项目
        经验主意预估模型（代码行(LOC)和功能点(FP)估算法等）
## 5. 确定时间进度
    1. 收集与进度相关的信息：总体工作量估算、人员数量、关键资源、项目时间安排等
    2. 确定各阶段任务安排和资源分配，确定里程碑
    3. 依据项目总体时间安排，形成进度计划
    4. 确定时间段
    5. 为每个测试目标规定合理的测试起始/中止时间。
    6. 通常情况下，功能性需求和非功能性需求的测试存在先后顺序，不能并行。
## 6. 评估风险
    1. 风险分析
    2. 对测试计划中所有要执行的内容进行潜在的风险分析并给出规避措施
    3. 确定项目中可能会出问题的地方
    4. 评估风险的发生概率
    5. 如风险发生后可能的影响程度
    6. 如何降低风险乃至避免风险的方法
## 7. 确定测试过程评估方法
    1. 确定测试过程评估方法
    2. 评估内容：测试工作进展/缺陷分布/质量评估
    3. 评估间隔：每天/周/月
    4. 评估人员/报告原则
